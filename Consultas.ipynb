```python
columnaKenny = dfKenny['Lines']
```

```python
frase_a_contar1Kenny = "dick"
```

```python
frase_a_contar2Kenny = "bitch"
```

```python
total1Kenny = columna.str.contains(frase_a_contar1Kenny, na=False).sum()
```

```python
total2Kenny = columna.str.contains(frase_a_contar2Kenny, na=False).sum()
```

```python
print(f'La palabra que mas se dice entre"{frase_a_contar1Kenny}" y "{frase_a_contar2Kenny}"aparece un total de {total1Kenny} la palabra"{frase_a_contar1Kenny}" y la palabra "{frase_a_contar2Kenny}" aparece un total de "{total2Kenny}" veces en la columna "{columnaKenny.name}".')
```

```python
plt.figure(figsize=(8, 6))
plt.bar(frase_a_contar1Kenny, total1Kenny, color='skyblue')
plt.bar(frase_a_contar2Kenny, total2Kenny, color='pink')
plt.title('Conteo de dick y bitch dichas por Kenny')
plt.xlabel('Frase')
plt.ylabel('Veces que se entiende cuando Kenny dice una de estas dos frases')
plt.ylim(0, total1Kenny + 5)
plt.ylim(0, total2Kenny + 5) 
plt.grid(axis='y')
plt.show()
```
